```python
from dotenv import load_dotenv

load_dotenv(override=True)
```

```python
with open("document.txt") as f:
    document = f.read()

print(document)
```

```python
from langchain_openai.embeddings import OpenAIEmbeddings

openai_embeddings = OpenAIEmbeddings()

print(openai_embeddings.embed_query("Hello, world!")[:10])
```

```python
from langchain_experimental.text_splitter import SemanticChunker

from rich import print

semantic_splitter = SemanticChunker(embeddings=openai_embeddings, breakpoint_threshold_type="interquartile")
docs = semantic_splitter.create_documents([document])

print(docs)
```

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai.embeddings import OpenAIEmbeddings
from rich import print

semantic_splitter = SemanticChunker(embeddings=openai_embeddings, breakpoint_threshold_type="percentile")
docs = semantic_splitter.create_documents([document])

print(docs)
```
